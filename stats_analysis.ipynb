```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
matches = pd.read_csv('matches.csv')
```

```python
matches['Tournament'].unique()
```

```python
tournaments = ['Primera Division', 'Clausura Argentina', 'Apertura Argentina', 'Copa Liga Profesional Argentina', 'Copa Libertadores', 'Copa Sudamericana', 'Primera B Nacional']
```

```python
def filter_by_tournament(tourney):
    tournaments = ['Primera Division', 'Clausura Argentina', 'Apertura Argentina', 'Copa Liga Profesional Argentina', 'Copa Libertadores', 'Copa Sudamericana', 'Primera B Nacional']
    flag = False
    for tournament in tournaments:
        if tourney == tournament:
            flag = True
    return flag
tournament_matches = matches['Tournament'].apply(lambda tourney: filter_by_tournament(tourney))
by_tourney = matches[tournament_matches]
by_tourney.reset_index(drop=True)
```

```python
end_index = by_tourney[by_tourney['Date']=='16 Nov 2012'].index.tolist()[0]
```

```python
by_date = by_tourney.iloc[:end_index,:]
by_date
```

```python
by_home = by_date[by_date['Home/Away']=='Home']
```

```python
by_home.reset_index(drop=True)
```

```python
start_covid_index = by_tourney[by_tourney['Date']=='01 Mar 2020'].index.tolist()[0]
```

```python
end_covid_index = by_tourney[by_tourney['Date']=='20 Sep 2021'].index.tolist()[0]
```

```python
upper_row = by_home.iloc[0:end_covid_index,:]
lower_row = by_home.iloc[start_covid_index:,:]
matches_df = pd.concat([upper_row,lower_row], axis=0, ignore_index=True)
matches_df = matches_df[matches_df['Opponent']!='Temperley']
matches_df.reset_index(drop=True)
```

```python
sns.countplot(x='Result',data=matches_df,hue='Tournament')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
